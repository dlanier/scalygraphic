### Equations that produce face-like figures with the random parameter generators used to find them
## nlC4Nr.m
```matlab
function count = nlC4Nr(Z,S_lim,B_lim,c)
    if nargin == 0
        count = rand(4,1) .* [0.1 2 2 0.2].';
        return
    end

    Zc = Z;                       count = 1;
    while count < S_lim && abs(Z-Zc) < B_lim
        Z = Z^(2*Z^(-c(1)^(Z^-c(2))^(Z^-c(3))^(Z^-c(4))));
        count = count + 1;
    end
```

## nlC7Nr.m
```matlab
function count = nlC7Nr(Z,S_lim,B_lim,c)
    if nargin == 0
        count =cumsum(cumsum(sort(rand(7,1)))).*[1 -1 1 -1 1 -1 1]';
        return
    end

    Zc = Z;                       count = 1;
    while count < S_lim && abs(Z-Zc) < B_lim
        Z = Z^(2*Z^(-c(1)^(Z^-c(2))^(Z^-c(3))^(Z^-c(4))^(Z^-c(5))^(Z^-c(6))^(Z^-c(7))));
        count = count + 1;
    end
```
## daPwrAFxNr.m
```matlab
function count = daPwrAFxNr(Z,S_lim,B_lim,c)
% variable number of input parameters - see parameter search below
    if nargin == 0
        count = -cumsum(abs(randn(3,1)));
        %count = [randn; randn(ceil(rand*3),1)];
        %count = cumsum(cumsum(sort(rand(ceil(rand*11),1))));
        return
    end
    
    nc = numel(c);
    
    Zc = Z;                           count = 1;
    while count < S_lim && abs(Z-Zc) < B_lim
        for n = 1:nc
            Z = 1/Z + -1^n * Z^( n*Z^(c(n)^n ));
        end
        count = count + 1;
    end
```
