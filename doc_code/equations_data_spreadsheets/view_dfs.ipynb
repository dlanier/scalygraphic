```python
eq_data_df = pd.read_csv('eq_data_Oct_17_2018.df', index_col=0, header=0, sep='\t').fillna('N/A')
print(eq_data_df.shape)
print(eq_data_df.index.name)
eq_data_df.head()
```

```python
from collections import defaultdict

#                                                 dictionary of uniqueness: various fields of dupe sets  -- 2 do
eq_str_list = list(eq_data_df['eq_string'].values)
unique_eq_strs = list(set(eq_str_list))

str_dict = {}
for idx, row in eq_data_df.iterrows():
    if row['eq_string'] in str_dict:
        str_dict[row['eq_string']] += 1
    else:
        str_dict[row['eq_string']] = 1

max_dupes = 0
histo_bins = {}
for k, v in str_dict.items():
    if v > 1:
        if v in histo_bins:
            histo_bins[v] += 1
        else:
            histo_bins[v] = 1
            
    if v > max_dupes:
        max_dupes = v

print('Distribution of duplicates')
for k in sorted(list(histo_bins.keys())):
    print(k, histo_bins[k])

print('max duplicates ', max_dupes)
print('Size of eq-string dictionary =',len(str_dict))

print('%i named equation m-files, of which\n%i are unique when considered as strings'%(len(eq_str_list), 
                                                                                      len(unique_eq_strs)))
print('\nNote: the while loop may modify the behavior of duplicate equations.')
print('Ergo total number of equations may be the unique number of Complex Iterated Function Systems')
```
